# **Question for the Assignment:**
# Two tangents AB and BC intersect at a point B at 150.5m chainage. Let us calculate all the necessary data for setting out a circular curve of 100m radius and deflection angle of 30 degrees using offset from long chord.

```python
import math
from sympy import *
```

```python
def_angle = float(input("Enter the deflection angle of the curve (integer/float values): "))
radius = float(input("Enter the radius of the curve (integer/float values): "))
chainage = (input("Enter the chainage value (integer/float values): "))
```

```python
#<-----Calculation Metrics----->
len_tangent = (radius*math.tan(def_angle*3.14/360)) #Since math.tan takes radians as input
len_curve = (2*3.14*radius*def_angle/360)
len_long_chord =(2*radius*math.sin(def_angle*3.14/360))
mid_ordinate = (radius*(1-math.cos(def_angle*3.14/360)))
apex_distance =(radius * (simplify(sec(def_angle*3.14/360))-1))
values_split = chainage.split('.') #Returns 240 and 55 separately as a list
value1= int(values_split[0])
dec='.'+values_split[-1] #Making 55 as 0.55
value2= float(dec)
chainage_at_v = float(chainage)
chainage_at_t1= chainage_at_v - len_tangent
chainage_at_t2 = chainage_at_t1+len_curve
#Intermediate Chord Length
def closest_greater_multiple_of_30(number):
    # Here we are calculating the closest multiple of 30 that is just greater than chainage at T1
    closest_multiple = ((number // 30) + 1) * 30
    return closest_multiple
chord_len_1=closest_greater_multiple_of_30 (chainage_at_t1) - chainage_at_t1
def closest_lower_multiple_of_30(number):
    # Here we are calculating the closest multiple of 30 that is just lower than chainage at T2
    closest_multiple = (number // 30) * 30
    return closest_multiple
chord_len_2=chainage_at_t2 - closest_lower_multiple_of_30 (chainage_at_t2)
other_chord_len = 30
no_of_int_chord = int(((len_curve - chord_len_1 - chord_len_2 )/other_chord_len)+2)
```

```python
#<-----Printing the Metrics----->
print("<---Calculated Values--->")
print(f" Radius: {radius}m \n Length of Tangent: {len_tangent}m \n Length of Curve: {len_curve}m \n Length of the Longest chord: {len_long_chord}m")
print(f" Mid-ordinate: {mid_ordinate}m \n Apex distance: {apex_distance}m \n Chainage at V: {chainage_at_v}m \n Chainage at T1/Point of Curve: {chainage_at_t1}m \n Chainage at T2/Point of Tangency: {chainage_at_t2}m \n First chord length: {chord_len_1}m \n Last chord length: {chord_len_2}m \n Number of intermediate chords: {no_of_int_chord}")
```
